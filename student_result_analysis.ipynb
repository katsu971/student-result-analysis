```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv("data/Expanded_data_with_more_features.csv")
print(df.head())
```

```python
print(df.describe())
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
#drop unnamed column
df=df.drop("Unnamed: 0", axis=1)
print(df.head())
```

```python
#gender distribution
plt.figure(figsize=(5,5))
ax=sns.countplot(data=df, x="Gender")
ax.bar_label(ax.containers[0])
plt.title("Gender Distribution")
plt.show()
```

```python
 #from the above chart we have analysed that: the number of females is more than males
```

```python
gb=df.groupby("ParentEduc").agg({"MathScore":'mean', "ReadingScore":'mean', "WritingScore":'mean'})
print(gb)
```

```python
plt.figure(figsize=(4,4))
sns.heatmap(gb, annot= True)
plt.title("Relationship between Parent Education and Scores")

plt.show()
```

```python
#from the above heatmap we can see that higher education of parents leads to better academic performance in kids
```

```python
gb1=df.groupby("ParentMaritalStatus").agg({"MathScore":'mean', "ReadingScore":'mean', "WritingScore":'mean'})
print(gb1)
```

```python
plt.figure(figsize=(4,4))
sns.heatmap(gb1, annot= True)
plt.title("Relationship between Parent Marital Status and Scores")

plt.show()
```

```python
#from the above heatmap we can incur that parent's marital status has negligible impact on student's scores
```

```python
#Distribution of Ethnic groups
groupA=df.loc[(df['EthnicGroup']=='group A')].count()
groupB=df.loc[(df['EthnicGroup']=='group B')].count()
groupC=df.loc[(df['EthnicGroup']=='group C')].count()
groupD=df.loc[(df['EthnicGroup']=='group D')].count()
groupE=df.loc[(df['EthnicGroup']=='group E')].count()
l=['groupA', 'groupB', 'groupC', 'groupD', 'groupE']
mlist=[groupA["EthnicGroup"], groupB["EthnicGroup"], groupC["EthnicGroup"], groupD["EthnicGroup"], groupE["EthnicGroup"]]
plt.pie(mlist, labels=l, autopct='%1.1f%%')
plt.title("Distribution of Ethnic Groups")
plt.show()
```
